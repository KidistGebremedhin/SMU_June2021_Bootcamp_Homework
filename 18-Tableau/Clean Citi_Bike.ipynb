```python
import pandas as pd
import numpy as np
```

```python
master_df = pd.DataFrame()

for i in range(1,13):
    month = str(i).zfill(2)
    df = pd.read_csv(f"Data/2019{month}-citibike-tripdata.csv", low_memory=False)    
    df = df.sample(frac=0.02, random_state = 42)
    master_df = master_df.append(df)
    print(i)
```

```python
master_df['Gender'] = master_df['gender'].replace([0,1,2],['unknown',"Male","Female"])
master_df = master_df.drop('gender',1) 
master_df.head()
```

```python
master_df= master_df.reset_index()
master_df = master_df.rename(columns={"index":"ID"})
master_df['ID'] = master_df.index + 1
master_df.head()
```

```python
master_df.to_csv("citibike_2019.csv", index=False)
```
